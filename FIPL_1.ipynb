1. PIL - Python Imaging Library



```python
from PIL import Image
import numpy as np #using numpy to convert image to arrays

```

```python
#Read image
img = Image.open("/content/drive/MyDrive/Dataset/agri_0_1017.jpeg") # image path
#Not a numpy array
print(type(img))
```

```python
#Output Images
img.show()
```

```python
#Printing format of image
print(img.format)
```

```python
#printing mode of image
print(img.mode)
```

```python
#PIL is not a default numpy array but can be used to convert into an array
img1 = np.asarray(img)
print(type(img1))
```

2. Scikit image

```python
from skimage import io , img_as_float , img_as_ubyte
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = img_as_float(io.imread("/content/drive/MyDrive/Dataset/Alloy_gradient.jpg"))
print(image)
print(type(img))
print(img.format)
print(img.mode)
```

```python
image8byte = img_as_ubyte(image) # image to byte conversion from float
print(image8byte)
```

```python
plt.imshow(image)
```

```python
plt.imshow(image8byte)
```

Difference is in storing the image not in displaying as float take more storage than integer (byte)

3 - Open CV

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
grey_img = cv2.imread("/content/drive/MyDrive/Dataset/Alloy_gradient.jpg")
color_img = cv2.imread("/content/drive/MyDrive/Dataset/raccoon.png")
```

```python
#image opened using cv2 are numpy arrays
print(type(grey_img))
print(type(color_img))
```

```python
plt.figure(figsize = (15,15))
plt.subplot(1,2,1) # creating multiple plots in single figure
plt.imshow(grey_img , 'gray')
plt.title('Gray image', fontsize = 14)
plt.axis('on')
plt.subplot(1,2,2)
plt.imshow(color_img)
plt.title("Color image" , fontsize=14)
plt.axis('off')


```
